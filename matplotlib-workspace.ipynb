```python
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50) * 2
y2 = np.random.rand(50) * 0.5
x3 = np.random.randn(50)
x4 = np.linspace(0, 10, 50)
y4 = np.sin(x4) + np.random.rand(50) * 0.5

axes[0, 0].scatter(x1, y1)
axes[0, 0].set_title('Scatter Plot 1')

axes[0, 1].scatter(x2, y2, color='red')
axes[0, 1].set_title('Scatter Plot 2')

axes[1, 0].plot(x3)
axes[1, 0].set_title('Scatter Plot 3')

axes[1, 1].scatter(x4, y4, color='green', alpha=0.7)
axes[1, 1].set_title('Scatter Plot 4')

plt.tight_layout()
plt.show()
```
